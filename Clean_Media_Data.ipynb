# First file to Clean

```python
df1 = pd.read_csv("./Media_Data/social media influencers - instagram sep-2022.csv") 
df1
```

```python
df1.columns
```

```python
# 1. Remove leading/trailing whitespace from column names
df1.columns = df1.columns.str.strip()

# 2. Standardize column names (lowercase and replace spaces with underscores)
df1.columns = df1.columns.str.lower().str.replace(' ', '_')

# 3. Clean string columns (if any) - assuming potential whitespace in these columns
string_columns = df1.select_dtypes(include='object').columns
for col in string_columns:
    df1[col] = df1[col].str.strip()
    
df1
```
