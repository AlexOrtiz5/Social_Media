```python
# 2. Identify and Handle Duplicate Columns
# Create a set of unique column names
unique_cols_insta = set()
duplicate_cols_insta = []
for col in merged_instagram_df.columns:
    if col in unique_cols_insta:
        duplicate_cols_insta.append(col)
    else:
        unique_cols_insta.add(col)

if duplicate_cols_insta:
    print(f"Duplicate columns found in merged_instagram_df: {duplicate_cols_insta}")
    # You might want to inspect these and decide how to handle them.
    # For now, let's keep the first occurrence.
    merged_instagram_df = merged_instagram_df.loc[:, ~merged_instagram_df.columns.duplicated()]
    print(f"Shape after removing duplicate columns: {merged_instagram_df.shape}")
```

```python
# 3. Handle Missing Values (NaN)
print(f"\nMissing values before handling:")
print(merged_instagram_df.isnull().sum().sort_values(ascending=False).head())

# Decide on a strategy for missing values. Options include:
# - Imputation (filling with mean, median, mode, or more advanced techniques)
# - Dropping rows with any NaN
# - Dropping columns with too many NaNs
# For now, let's drop rows where key identifier (e.g., instagram_name) is missing
merged_instagram_df.dropna(subset=['instagram_name'], inplace=True)
merged_instagram_df.fillna({'category_1': 'Unknown', 'category_2': 'Unknown'}, inplace=True) # Fill categorical NaNs
# For numerical columns, you might impute or drop based on context
numeric_cols_insta = merged_instagram_df.select_dtypes(include=['number']).columns
for col in numeric_cols_insta:
    if merged_instagram_df[col].isnull().sum() > 0:
        # Example imputation with median (can be replaced with other strategies)
        median_val = merged_instagram_df[col].median()
        merged_instagram_df[col].fillna(median_val, inplace=True)
        print(f"Filled missing values in '{col}' with median: {median_val}")

print(f"Missing values after handling:")
print(merged_instagram_df.isnull().sum().sort_values(ascending=False).head())
print(f"Shape after handling missing values: {merged_instagram_df.shape}")
```
