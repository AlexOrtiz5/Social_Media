# Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read instagram
insta = pd.read_csv('./Clean_Data/merged_instagram_data_cleaned.csv')

# read tiktok
tiktok = pd.read_csv('./Clean_Data/merged_tiktok_data_cleaned.csv')

# read youtube
youtube = pd.read_csv('./Clean_Data/merged_youtube_data_cleaned.csv')
```

```python
# --- Analysis for merged_instagram_df ---
print("\n--- Analyzing merged_instagram_df ---")

# Basic Descriptive Statistics
print("\nDescriptive Statistics:")
print(insta.describe(include='all'))

# Top 10 by Subscribers
if 'subscribers' in insta.columns:
    top_10_insta_subscribers = insta.sort_values(by='subscribers', ascending=False).head(10)
    print("\nTop 10 by Subscribers:")
    print(top_10_insta_subscribers[['instagram_name', 'subscribers']].fillna('N/A'))

    if 'subscribers' in insta.columns:
        top_10_insta_subscribers = insta.sort_values(by='subscribers', ascending=False).head(10)
        print("\nTop 10 by Subscribers:")
        print(top_10_insta_subscribers[['instagram_name', 'subscribers']].fillna('N/A'))

        plt.figure(figsize=(10, 6))
        sns.barplot(x='subscribers', y=top_10_insta_subscribers['instagram_name'].fillna('N/A'), data=top_10_insta_subscribers, palette='viridis') # Explicitly pass data
        plt.title('Top 10 Instagram by Subscribers')
        plt.xlabel('Subscribers')
        plt.ylabel('Instagram Name')
        plt.tight_layout()
        plt.show()

# Distribution of Subscribers
if 'subscribers' in insta.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(insta['subscribers'], bins=50, kde=True)
    plt.title('Distribution of Instagram Subscribers')
    plt.xlabel('Subscribers')
    plt.ylabel('Frequency')
    plt.show()

# Average Engagement
engagement_cols_insta = [col for col in insta.columns if 'engagement' in col]
if engagement_cols_insta:
    print("\nAverage Engagement Metrics:")
    print(insta[engagement_cols_insta].mean())

    plt.figure(figsize=(8, 5))
    insta[engagement_cols_insta].mean().plot(kind='bar')
    plt.title('Average Engagement Metrics on Instagram')
    plt.ylabel('Average Value')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Top Categories (Category 1)
if 'category_1' in insta.columns and insta['category_1'].nunique() > 1:
    top_insta_categories = insta['category_1'].value_counts().nlargest(10)
    print("\nTop 10 Category 1:")
    print(top_insta_categories)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_insta_categories.index, y=top_insta_categories.values, palette='plasma')
    plt.title('Top 10 Instagram Category 1')
    plt.xlabel('Category')
    plt.ylabel('Number of Influencers')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
print("\n--- Analyzing merged_tiktok_df ---")

# Basic Descriptive Statistics
print("\nDescriptive Statistics:")
print(tiktok.describe(include='all'))

# Top 10 by Subscribers
if 'subscribers' in tiktok.columns:
    top_10_tiktok_subscribers = tiktok.sort_values(by='subscribers', ascending=False).head(10)
    print("\nTop 10 by Subscribers:")
    print(top_10_tiktok_subscribers[['tiktok_name', 'subscribers']].fillna('N/A'))

    plt.figure(figsize=(10, 6))
    sns.barplot(x='subscribers', y=top_10_tiktok_subscribers['tiktok_name'].fillna('N/A'), palette='viridis')
    plt.title('Top 10 TikTok by Subscribers')
    plt.xlabel('Subscribers')
    plt.ylabel('TikTok Name')
    plt.tight_layout()
    plt.show()

# Distribution of Subscribers
if 'subscribers' in tiktok.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(tiktok['subscribers'], bins=50, kde=True)
    plt.title('Distribution of TikTok Subscribers')
    plt.xlabel('Subscribers')
    plt.ylabel('Frequency')
    plt.show()

# Average Engagement
engagement_cols_tiktok = [col for col in tiktok.columns if 'engagement' in col]
if engagement_cols_tiktok:
    print("\nAverage Engagement Metrics:")
    print(tiktok[engagement_cols_tiktok].mean())
    
    plt.figure(figsize=(8, 5))
    tiktok[engagement_cols_tiktok].mean().plot(kind='bar')
    plt.title('Average Engagement Metrics on TikTok')
    plt.ylabel('Average Value')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
