# Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read instagram
insta = pd.read_csv('./Clean_Data/merged_instagram_data_cleaned.csv')

# read tiktok
tiktok = pd.read_csv('./Clean_Data/merged_tiktok_data_cleaned.csv')

# read youtube
youtube = pd.read_csv('./Clean_Data/merged_youtube_data_cleaned.csv')
```

```python
# --- Analysis for merged_instagram_df ---
print("\n--- Analyzing merged_instagram_df ---")

# Basic Descriptive Statistics
print("\nDescriptive Statistics:")
print(insta.describe(include='all'))
```

```python
# Top Categories (Category 1)
if 'category_1' in insta.columns and insta['category_1'].nunique() > 1:
    top_insta_categories = insta['category_1'].value_counts().nlargest(10)
    print("\nTop 10 Category 1:")
    print(top_insta_categories)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_insta_categories.index, y=top_insta_categories.values, palette='plasma')
    plt.title('Top 10 Instagram Category 1')
    plt.xlabel('Category')
    plt.ylabel('Number of Influencers')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
print("\n--- Analyzing merged_tiktok_df ---")

# Basic Descriptive Statistics
print("\nDescriptive Statistics:")
print(tiktok.describe(include='all'))
```

```python
# Average Engagement
engagement_cols_tiktok = [col for col in tiktok.columns if 'engagement' in col]
if engagement_cols_tiktok:
    print("\nAverage Engagement Metrics:")
    print(tiktok[engagement_cols_tiktok].mean())
    
    plt.figure(figsize=(8, 5))
    tiktok[engagement_cols_tiktok].mean().plot(kind='bar')
    plt.title('Average Engagement Metrics on TikTok')
    plt.ylabel('Average Value')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
print(f"\n--- Analyzing merged_youtube_df ---")

# Basic Descriptive Statistics
print("\nDescriptive Statistics:")
print(youtube.describe(include='all'))
```

```python
# Average Engagement
engagement_cols = [col for col in youtube.columns if 'avg' in col or 'views_' in col or 'likes_' in col or 'comments_' in col]
if engagement_cols:
    print("\nAverage Engagement Metrics:")
    print(youtube[engagement_cols].mean())

    plt.figure(figsize=(8, 5))
    youtube[engagement_cols].mean().plot(kind='bar')
    plt.title('Average Engagement Metrics on YouTube')
    plt.ylabel('Average Value')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
